```python
#import library name as Pandas used to tables 
import pandas as pd
#we have dump from customer that need to  read file name called read_csv
dataset=pd.read_csv("Book1.csv") 
```

```python
 #To Cross check o/p dataset split
dataset
```

```python
independent=dataset[["Years_of_experience"]]
#bascially we need to intiate variable so RHS to LHS value & To Cross check o/p dataset split
independent
```

```python
dependent=dataset[["Salary"]] #bascially we need to intiate variable so RHS to LHS value 
dependent# To Cross check o/p dataset split 
```

```python
from sklearn.model_selection import train_test_split #use one of library name called sklearn to bring up model selection compare to train to test set
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)#intiate variable X&Y i/p & o/p 
```

```python
from sklearn.linear_model import LinearRegression#library name called sklearn to bring up linear model regression method 
regressor=LinearRegression() #we need to intiate variable so RHS to LHS value
regressor.fit(x_train,y_train)#.fit , call to action model is ready to perform
```

```python
weight=regressor.coef_ #to find the weight function name called .coef_
weight #to run the weight 
```

```python
bias=regressor.intercept_ #to find the bias function name called .intercept_
bias #to run the bias
```

```python
y_pred=regressor.predict(x_test) #now the Model is working fine or not 
from sklearn.metrics import r2_score #use library fun to import r2 to check 0 OR 1 
r_score=r2_score(y_test,y_pred) # to predict the model name called as test 
r_score #test o/p
```

```python
import pickle # to save model called as Pickle libiary
filename="finalized_model_linear.sav" # save the model file name called .sav
pickle.dump(regressor,open(filename, 'wb')) #to intiate save model here to run write binary 
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb')) #to intiate save model here to run read binary 
```
